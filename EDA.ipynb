## Importing necessary libraries

```python
import json
import pandas as pd
```

## Reading the data

```python
# read the json file data/dataset.json and convert it to a pandas dataframe
def read_json():
    with open("Data/dataset.json", "r") as read_file:
        data = json.load(read_file)

    df = pd.DataFrame.from_dict(data, orient='index')
    return df

df = read_json()
df.head(10)
```

```python
# creating 3 subcolumns under annotators column named as annotator1, annotator2, annotator3
def create_annotator_subcolumns(df):
    df['annotator1'] = df['annotators'].apply(lambda x: x[0])
    df['annotator2'] = df['annotators'].apply(lambda x: x[1])
    df['annotator3'] = df['annotators'].apply(lambda x: x[2])
    return df

df = create_annotator_subcolumns(df)
df.drop('annotators', axis=1, inplace=True)
df.head(10)
```

```python
# creating 3 rationales as ratoinale1, rationale2, rationale3
# if the list is empty, then set each rationale to empty list
def create_rationale_subcolumns(df):
    df['rationale1'] = df['rationales'].apply(lambda x: x[0] if len(x) > 0 else [])
    df['rationale2'] = df['rationales'].apply(lambda x: x[1] if len(x) > 1 else [])
    df['rationale3'] = df['rationales'].apply(lambda x: x[2] if len(x) > 2 else [])
    return df

df = create_rationale_subcolumns(df)
df.drop('rationales', axis=1, inplace=True)
df.head(10)

```

```python
# save the dataframe to a excel file
df.to_excel("dataset.xlsx")
```

## EDA

```python
# get the unique targets from the dataset
def get_unique_targets(df):
    unique_targets = set()
    for i in range(0, len(df)):
        unique_targets = list(unique_targets)
        unique_targets.extend(df['annotator1'][i]['target'])
        unique_targets.extend(df['annotator2'][i]['target'])
        unique_targets.extend(df['annotator3'][i]['target'])
        unique_targets = set(unique_targets)
    return list(unique_targets)

unique_targets = get_unique_targets(df)

print(sorted(unique_targets))
print(len(unique_targets))
```
